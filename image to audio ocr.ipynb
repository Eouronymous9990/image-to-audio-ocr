```python
def image_to_string (img):
    pytesseract.pytesseract.tesseract_cmd = r"E:\vsc\tesseract.exe"#put your tesseract path 
    text=pytesseract.image_to_string(img,lang="eng")
    
    print(text)  

    tts = gTTS(text)
    tts.save('english audio.mp3')

    
    

image_to_string(img) 
    
audio_path='english audio.mp3'
ipd.Audio(audio_path, autoplay=True)  
```
